```python
%config InlineBackend.figure_format = 'retina'
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('/Users/jgreco/Desktop/puppy.jpg')
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img)
#plt.axis('off');
```

```python
reds = img.copy()
reds[:, :, 1] = 0
reds[:, :, 2] = 0

greens = img.copy()
greens[:, :, 0] = 0
greens[:, :, 2] = 0

blues = img.copy()
blues[:, :, 0] = 0
blues[:, :, 1] = 0


fig, axes = plt.subplots(1, 4, figsize=(16, 8), 
                         subplot_kw=dict(xticks=[], yticks=[]))

fs = 18
axes[0].imshow(img)
axes[0].set_title('full color puppy', fontsize=fs)
axes[1].imshow(reds)
axes[1].set_title('red channel', fontsize=fs)
axes[2].imshow(greens)
axes[2].set_title('green channel', fontsize=fs)
axes[3].imshow(blues)
axes[3].set_title('blue channel', fontsize=fs);
```
